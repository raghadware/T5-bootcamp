# Rain Prediction - Australia

in this notebook we try to predict next-day rain by training classification models on the target variable RainTomorrow.

#### **Dataset**:
> [Rain in Australia - kaggle](https://www.kaggle.com/datasets/jsphyg/weather-dataset-rattle-package)
> This dataset contains about 10 years of daily weather observations from many locations across Australia.


**Import libraries**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn import model_selection
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
weather = pd.read_csv('weatherAUS.csv')
```

```python
weather
```

#### *Preprocessing*

> - Lowercase column labels.
> - Replace the values in the `raintoday` and `raintomorrow` columns to 0 for No and 1 for Yes.
> - convert data into numbers using `LabelEncoder`
> - Fill missing values with previous values , then fill the rest with next value.
> - drop unnecessary columns
> - split to train and test data  

```python
weather.info()
```

```python
weather.describe()
```

```python
weather['WindGustDir'].unique()
```

```python
weather[weather['WindGustDir'].isna() == True][['WindGustDir','WindDir9am','WindDir3pm']]
```

```python
#change column labels to lower case
weather = weather.rename(columns=str.lower)
```

**Replace the values in the `RainToday` and `Raintomorrow` columns to 0 for No and 1 for Yes**

```python
weather['raintoday_num'] = pd.get_dummies(weather['raintoday'] , dummy_na=False)['Yes']
weather['raintomorrow_num'] = pd.get_dummies(weather['raintomorrow'] , dummy_na=False)['Yes']
```

```python
#Drop unnecessary columns
weather.drop(['raintoday','raintomorrow' , 'sunshine' , 'evaporation','rainfall' , 'date'] , axis=1 , inplace = True)
```

```python
#rename columns
weather = weather.rename(columns={'raintoday_num':'raintoday','raintomorrow_num':'raintomorrow'})
```

**Use `LabelEncoder` to convert values to numerical and prepare for modeling**

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

to_encode = ['location','winddir9am','winddir3pm' , 'windgustdir'] 

for label in to_encode:
    weather[label] = le.fit_transform(weather[label])
```

```python
#fill all columns with the previous values
weather = weather.fillna(method='ffill')
#fill the rest of the columns with the next velue
weather = weather.fillna(method='bfill')
```

```python
weather.isna().sum()
```

#### Look for correlations

```python
corr = weather.corr()
plt.figure(figsize=(13,13))
sns.heatmap(corr, annot=True, fmt='.2f',  vmin=-1 , vmax=1 , cmap = 'crest')
plt.show()
```

# Rain Prediction

**Separate X from Y**

```python
X = weather.drop('raintomorrow' , axis=1)
Y = weather['raintomorrow']
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(X , Y  , test_size = 0.2)
X_train , X_val , Y_train , Y_val = train_test_split(X_train , Y_train  , test_size = 0.2)
```

## Modeling

### Ensembling

```python
knn_classifier = KNeighborsClassifier()
tree_classifier = DecisionTreeClassifier()
logistic_classifier = LogisticRegression(max_iter = 1000)

# Create an ensemble model using VotingClassifier
estimators = [('knn', knn_classifier), ('tree', tree_classifier), ('logistic', logistic_classifier)]
ensemble = VotingClassifier(estimators, voting='hard')

# Fit the ensemble model on the training data
ensemble.fit(X_train, Y_train)

# Evaluate the ensemble model
accuracy = ensemble.score(X_test, Y_test)
print(f'Accuracy of the ensemble model: {accuracy:.2f}')
```

**The model predicts the rain tomorrow with %85 accuracy**
