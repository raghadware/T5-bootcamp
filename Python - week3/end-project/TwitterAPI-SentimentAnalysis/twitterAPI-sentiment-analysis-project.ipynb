#### Getting `full_text` from `tweets` json

- **create empty list `tweets_list` to hold tweets `full_text`**
- **create function to find `full_text` from every tweet **
- **check if the data sent to the function is in `dict` type or `list` type**
- **if `data` a dictionary loop through every key and value among that dictionary items and find values with key `full_text` append tha value to the `tweets_list`, else send the value to call that function again**
- **if `data` a list loop through every item in the `data` list and call `find_full_text()` function again for that item**
- **this process continues to get all tweets `full_text`**

```python
tweets_list=[]

def find_full_text(data):
    if isinstance(data, dict):
        for key, value in data.items():
            if key == "full_text":
                tweets_list.append(value) 
            else:
                find_full_text(value)
    elif isinstance(data, list):
        for item in data:
            find_full_text(item)
```

**Call function and send tweets**

```python
find_full_text(tweets)
```

**After calling the function we print the final list that contains all tweets**

```python
tweets_list
```

**Create a datafram `tweets_df` with `Tweet` column**

```python
tweets_df = pd.DataFrame(tweets_list, columns=['Tweet'])
```

**Clean tweets get rid of any `-\n` or `.` or `\n`**

```python
ch = ['-\n', '.', '\n']
#freplace every character with nothing 
for i in ch:
    tweets_df['Tweet'] = tweets_df['Tweet'].str.replace(i, '')
```

**Cut all hashtags from tweets as they won't be needed in the sentiment analysis**

```python
tweets_df['Tweet'] = tweets_df['Tweet'].str.split('#').str[0]
tweets_df
```

- **create a list to store the sentiment labels of each tweet in the dataframe `df`**
- **create `blob` object**
- **use the blob object to call the`.sentiment.polarity` function and store the numeric value in the `sentiment_score`**
- **used chained conditional statements to decide the labels and append them to the `sentiment_list`**
- **add the `sentiment_list` as a column in the `df['label']` already existing column**

```python
tweets_labels = []
for tweet in tweets_df['Tweet']:
    blob = TextBlob(tweet)
    tweet_score = blob.sentiment.polarity

    if tweet_score > 0:
        tweets_labels.append("Positive")
    elif tweet_score < 0:
        tweets_labels.append("Negative")
    else:
        tweets_labels.append("Neutral")

tweets_df['label'] = tweets_labels

tweets_df
```

**Save the final analysis result in csv file**

```python
tweets_df.to_csv('tweetS-sentiment-analysis.csv', index=False)
```
